Теперь оставим признаки только с кореляцией выше 0.05 с целевой переменной. На попарную корелляцию признаков смысла проверять нет, так как PCA гаранитрует их независимость.

## Нейросеть 

```python
import torch
import torch.nn as nn
```

Теперь отойдем от классических моделей машинного обучения и построим нейронную сеть.

Возьмем три линейных слоя с функцией активации гиперболического тангенса.

```python
class Net(nn.Module):
    
    def __init__(self):
        super(Net, self).__init__()
        self.layer_1 = nn.Linear(17, 60)
        self.layer_2 = nn.Linear(60, 60)
        self.layer_3 = nn.Linear(60, 1)
        self.f = nn.Tanh()
    
    def forward(self, x):
        x = self.layer_1(x)
        x = self.f(x)
        x = self.layer_2(x)
        x = self.f(x)
        x = self.layer_3(x)
        x = self.f(x)
        return x
```

```python
model = Net()
```

```python
X_train = torch.tensor(np.array(X_train), dtype=torch.float)
y_train = torch.tensor(np.array(y_train), dtype=torch.float)
```

Теперь обучим эту нейросеть

```python
loss_function = nn.MSELoss() 
optimizer = torch.optim.Adam(model.parameters(),lr = 0.001)

num_epochs = 900
batch_size = 200

for epoch in range(num_epochs):
    
    sum_loss = 0.0
    X_test_ = torch.tensor(np.array(X_test), dtype=torch.float)
    y_predict = model(X_test_).detach().numpy()
    y_test = np.array(y_test)
    
    for i in range(0, X_train.shape[0], batch_size):
        
        len_of_batch = min(X_train.shape[0] - i, batch_size)
        input_data = X_train[i : i + len_of_batch]
        labels = y_train[i : i + len_of_batch]
        
        output_data  = model(input_data)
        loss = loss_function(output_data, labels)
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        sum_loss = sum_loss + loss.item() * batch_size
        

    print("Epoch: {} - Loss:{:.8f}".format(epoch+1, sum_loss/X_train.shape[0] ))
```

```python
X_test_ = torch.tensor(np.array(X_test), dtype=torch.float)
y_predict = model(X_test_).detach().numpy()
print(r2_score(y_predict, y_test))
```

```python
plt.scatter(y_test, y_predict, alpha=0.5)
plt.title('predict vs real')
plt.xlabel('real')
plt.ylabel('predict')

plt.plot([0.0, 1.0], [0.0, 1.0], color='red')

plt.show()
```

Нейросеть имеет лучший прогноз в среднем!!! Выбросов и аномальных точек нет!!!

```python
if ((abs(y_test - y_predict) > 0.01).mean() == 0.0):
    print("Прогноз нейросети имеет отклонение не более 1%!!!")
    
if ((abs(y_test - y_predict) > 0.006).mean() < 0.01):
    print("99% прогнозов имеют отклонение не больше 0,6%!!!")
```
